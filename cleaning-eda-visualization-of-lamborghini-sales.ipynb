##**Data Cleaning**

###**Standardizing column names**

```python
lamborghini.columns = lamborghini.columns.str.strip().str.lower().str.replace(" ","_")
lamborghini.head()
```

###**Renaming column names**

```python
lamborghini = lamborghini.rename(columns={lamborghini.columns[5]:"base_price_in_usd",lamborghini.columns[8]:"has_turbo_engine"})
lamborghini.head()
```

###**Standardizing data in columns**

```python
lamborghini["model"] = lamborghini["model"].str.replace("Huracán (Temerario)","Temerario")
lamborghini["model"].value_counts()
```

##**Data Visualization**

###**Univariate analysis**

####**Distribution of base price of Lamborghini (USD)**

```python
f,ax = plt.subplots(figsize=(12,6))
ax.hist(lamborghini["base_price_in_usd"],bins=50,color="goldenrod",edgecolor="k")
ax.set_xlabel("Base price (USD)")
ax.set_ylabel("Count")
ax.set_title("Distribution of base price of Lamborghini (USD)")
plt.savefig("/kaggle/working/plots/distribution_of_base_price_of_lamborghini.png")
```

####**Insights:**
- The base price of Lamborghini cars ranges between 200000 to 590000 USD.
- Most of the Lamborghini cars are priced between 250000 to 280000 USD, and between 550000 to 575000 USD.

####**Distribution of horsepower in Lamborghini cars (hp)**

```python
f, ax = plt.subplots()
ax.boxplot(lamborghini["horsepower"])
ax.set_ylabel("Horsepower (hp)")
ax.set_xticklabels(["Horsepower"])
ax.set_title("Distribution of horsepower in Lamborghini cars (hp)")
plt.savefig("/kaggle/working/plots/distribution_of_horsepower_in_Lamborghini_cars_hp.png")
```

####**Insights:**
- Most of the horsepower in the Lamborghini cars are around 650 hp.
- The usual spread of horsepower is between 625 to 850 hp.
- The presence of outlier above the upper whisker denotes that few Lamborghini cars have a horsepower of 1000.

####**Distribution of sales volume**

```python
plt.hist(lamborghini["sales_volume"],color="blue",bins=20)
plt.xlabel("Sales volume")
plt.ylabel("Count")
plt.title("Distribution of sales volume")
plt.savefig("/kaggle/working/plots/distribution_of_sales_volume.png")
```

####**Insights:**
- The distribution of sales volume of Lamborghini cars are between 50 to 6300.

####**Percentage of Lamborghini cars sold each year by color**

```python
car_colors_map = {"Red":"red","Black":"black","White":"white","Yellow":"yellow","Orange":"darkorange","Silver":"lightslategrey","Blue":"#39BFFE","Green":"green","Grey":"grey"}
cars_sold_in_2020_color = lamborghini[lamborghini["year"]==2020]["color"].value_counts()
cars_sold_in_2021_color = lamborghini[lamborghini["year"]==2021]["color"].value_counts()
cars_sold_in_2022_color = lamborghini[lamborghini["year"]==2022]["color"].value_counts()
cars_sold_in_2023_color = lamborghini[lamborghini["year"]==2023]["color"].value_counts()
cars_sold_in_2024_color = lamborghini[lamborghini["year"]==2024]["color"].value_counts()
cars_sold_in_2025_color = lamborghini[lamborghini["year"]==2025]["color"].value_counts()
f,ax = plt.subplots(2,3,figsize=(14,8))
ax[0,0].pie(cars_sold_in_2020_color.values,labels=cars_sold_in_2020_color.index,explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2020_color.index],colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2020_color.index],wedgeprops={'edgecolor':'k'},autopct="%1.1f%%",pctdistance=0.77)
ax[0,0].set_title("Cars sold in 2020 by color (%)")
ax[0,1].pie(cars_sold_in_2021_color.values,labels=cars_sold_in_2021_color.index,colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2021_color.index],explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2021_color.index],wedgeprops={'edgecolor': 'k'},autopct="%1.1f%%",pctdistance=0.77)
ax[0,1].set_title("Cars sold in 2021 by color (%)")
ax[0,2].pie(cars_sold_in_2022_color.values,labels=cars_sold_in_2022_color.index,explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2022_color.index],colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2022_color.index],wedgeprops={"edgecolor":"k"},autopct="%1.1f%%",pctdistance=0.77)
ax[0,2].set_title("Cars sold in 2022 by color (%)")
ax[1,0].pie(cars_sold_in_2023_color.values,labels=cars_sold_in_2023_color.index,explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2023_color.index],colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2023_color.index],wedgeprops={'edgecolor': 'k'},autopct="%1.1f%%",pctdistance=0.77)
ax[1,0].set_title("Cars sold in 2023 by color (%)")
ax[1,1].pie(cars_sold_in_2024_color.values,labels=cars_sold_in_2024_color.index,explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2024_color.index],colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2024_color.index],wedgeprops={'edgecolor': 'k'},autopct="%1.1f%%",pctdistance=0.77)
ax[1,1].set_title("Cars sold in 2024 by color (%)")
ax[1,2].pie(cars_sold_in_2025_color.values,labels=cars_sold_in_2025_color.index,explode=[0.10 if car_color=="Black" else 0 for car_color in cars_sold_in_2025_color.index],colors=[car_colors_map.get(car_color) for car_color in cars_sold_in_2025_color.index],wedgeprops={'edgecolor': 'k'},autopct="%1.1f%%",pctdistance=0.77)
ax[1,2].set_title("Cars sold in 2025 by color (%)")
f.suptitle("Percentage of Lamborghini cars sold each year by color(%)")
plt.savefig("/kaggle/working/plots/percentage_of_lamborghini_cars_sold_each_year_by_color.png");
```

####**Insights:**
- Black color Lamborghini cars are the most sold in every year from 2020 to 2025.
- White color stays in the second place for the most sold Lamborghini cars.
- Orange color is the least sold color of Lamborghinis in 2020, 2022, 2023, 2024 and 2025.
- Green color Lamborghini cars are the fourth most sold colors of Lamborghini in the year 2024.

####**Sales volume of Lamborghini cars by region**

```python
f, ax = plt.subplots()
ax.boxplot([lamborghini[lamborghini["region"]=="Americas"]["sales_volume"],lamborghini[lamborghini["region"]=="EMEA"]["sales_volume"],lamborghini[lamborghini["region"]=="APAC"]["sales_volume"]])
ax.set_ylabel("Sales volume")
ax.set_xticklabels(["Americas","EMEA","APAC"])
ax.set_title("Sales volume of Lamborghini cars by region")
plt.savefig("/kaggle/working/plots/sales_volume_of_lamborghini_cars_by_region.png");
```

####**Number of hybrid cars sold by model name**

```python
gasoline_cars_sold = lamborghini[lamborghini["fuel_type"]=="Gasoline"].groupby("model")["sales_volume"].sum()
hybrid_cars_sold = lamborghini[lamborghini["fuel_type"]=="Hybrid"].groupby("model")["sales_volume"].sum()
f,ax = plt.subplots(1,2,figsize=(14,6))
ax[0].bar(gasoline_cars_sold.index,gasoline_cars_sold.values,width=0.35,color="orangered")
ax[0].set_xlabel("Model name")
ax[0].set_ylabel("Number of cars sold")
ax[0].set_title("Number of gasoline cars sold by model name")
ax[1].bar(hybrid_cars_sold.index,hybrid_cars_sold.values,width=0.35,color="lawngreen")
ax[1].set_xlabel("Model name")
ax[1].set_ylabel("Number of cars sold")
ax[1].set_title("Number of hybrid cars sold by model name")
plt.savefig("/kaggle/working/plots/number_of_hybrid_cars_sold_by_model_name.png");
```

####**Insights:**
- Lamborghini Urus ranks top as the most sold car running on gasoline with the number of cars around 55000.
- Lamborghini Aventador ranks as the least sold car running on gasoline with the number of cars sold as around 6500.
- Lamborghini URUS SE tops as the most sold car running on hybrid engine with the number of cars sold between 27000 to 29000.
- Lamborghini Temerario ranks last among the cars running on hybrid engine with the number of cars sold as around 4000.

####**Comparison of Lamborghini car models sold between 2020 to 2025**

```python
f, ax = plt.subplots(2,3,figsize=(13,7))
years = lamborghini["year"].unique()
colors_year = ["crimson","teal","lime",'darkorange',"magenta","royalblue"]
row_num=0
col_num=0
year_ptr = 0
for row_num in range(2):
  for col_num in range(3):
    ax[row_num,col_num].bar(lamborghini[(lamborghini["year"]==years[year_ptr])].groupby("model")["sales_volume"].sum().index,lamborghini[(lamborghini["year"]==years[year_ptr])].groupby("model")["sales_volume"].sum().values,width=0.27,color=colors_year[year_ptr])
    ax[row_num,col_num].set_xlabel("Model name")
    ax[row_num,col_num].set_ylabel("Number of cars sold")
    ax[row_num,col_num].set_title(f"Number of cars sold in {years[year_ptr]} by model")
    year_ptr += 1
f.suptitle("Comparison of Lamborghini car models sold between 2020 to 2025")
f.tight_layout()
plt.savefig("/kaggle/working/plots/comparison_of_lamborghini_car_models_sold_between_2020_to_2025.png");
```

####**Insights:**
- Lamborghini Urus is the most sold car model by ranking in the first place between 2020 and 2023.
- From 2024 onwards, Lamborghini Urus SE has taken the first place as the most sold car.
- Lamborghini Aventador has remained in the third place continuously for three years from 2020 to 2022.
- The sales of Lamborghini Aventador has dipped less than 500 number of cars in the year 2023.
- Lamborghini Huracán is the second most sold car in 2024.
- Lamborghini Urus SE ranks first place among the number of cars sold in 2024 and 2025 with 14000 and 12000 cars respectively.
